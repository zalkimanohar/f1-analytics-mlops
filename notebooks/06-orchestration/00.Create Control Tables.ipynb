```python
# ============================================================
# CREATE CONTROL TABLES — LOCAL EXECUTION
# ============================================================

import os
import pandas as pd
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql import functions as F
import nbformat
from nbconvert import PythonExporter

# Load environment
def run_nb(path):
    with open(path, "r") as f:
        nb = nbformat.read(f, as_version=4)
    src, _ = PythonExporter().from_notebook_node(nb)
    exec(src, globals())

run_nb("/Users/manoharazalki/F1-Analytics/notebooks/00-common/01_environment_config.ipynb")
```

```python
# -----------------------------------------
# 1. Create control folder
# -----------------------------------------
control_path = f"{CONTROL_PATH}/batch_control"
os.makedirs(control_path, exist_ok=True)
```

```python
# -----------------------------------------
# 2. Create empty control table if not exists
# -----------------------------------------
control_file = f"{control_path}/data.parquet"

if not os.path.exists(control_file):
    empty_schema = StructType([
        StructField("batch_id", StringType(), True),
        StructField("status", StringType(), True),
        StructField("created_timestamp", TimestampType(), True),
        StructField("updated_timestamp", TimestampType(), True),
    ])
    spark.createDataFrame([], empty_schema).write.mode("overwrite").parquet(control_file)
    print("✔ Created empty control table")
else:
    print("✔ Control table already exists")

display(spark.read.parquet(control_file))
```
