```python
# ============================================================
# ENVIRONMENT CONFIG — LOCAL EXECUTION
# ============================================================

import os
from pyspark.sql import SparkSession
```

```python
# -----------------------------------------
# 1. Initialize Spark Session (Local Mode)
# -----------------------------------------
spark = (
    SparkSession.builder
        .appName("F1-Analytics-Local")
        .master("local[*]")
        .config("spark.sql.shuffle.partitions", "8")
        .config("spark.driver.memory", "4g")
        .getOrCreate()
)
```

```python
# -----------------------------------------
# 3. Base Paths
# -----------------------------------------
BASE_PATH = "/Users/manoharazalki/F1-Analytics/data"
```

```python
# -----------------------------------------
# 4. Define All Layer Paths
# -----------------------------------------
# Landing zone (raw files)
LANDING_PATH = f"{BASE_PATH}/landing"

# Bronze layer (raw → structured)
BRONZE_PATH = f"{BASE_PATH}/bronze"

# Silver layer (cleaned & standardized)
SILVER_PATH = f"{BASE_PATH}/silver"

# Gold layer (star schema)
GOLD_PATH = f"{BASE_PATH}/gold"

# Incremental landing zone
INCREMENTAL_PATH = f"{BASE_PATH}/incremental"

# Control tables (batch events)
CONTROL_PATH = f"{BASE_PATH}/control"
```

```python
# -----------------------------------------
# 5. Create Directories If Missing
# -----------------------------------------

paths = [
    LANDING_PATH,
    BRONZE_PATH,
    SILVER_PATH,
    GOLD_PATH,
    INCREMENTAL_PATH,
    CONTROL_PATH
]

for p in paths:
    if not os.path.exists(p):
        os.makedirs(p, exist_ok=True)
```

```python
# ----------------------------------------------
# 6. Helper: Build Full Paths for Each Dataset
# ----------------------------------------------

def bronze_path(table_name: str):
    return f"{BRONZE_PATH}/{table_name}"

def silver_path(table_name: str):
    return f"{SILVER_PATH}/{table_name}"

def gold_path(table_name: str):
    return f"{GOLD_PATH}/{table_name}"

def incremental_path(batch_folder: str):
    return f"{INCREMENTAL_PATH}/{batch_folder}"
```

```python
# ----------------------------------------------
# 7. Batch ID Generator
# ----------------------------------------------
from datetime import datetime

def generate_batch_id():
    return datetime.now().strftime("%Y%m%d_%H%M%S")
```

```python
# ----------------------------------------------
# 8. Print Configuration Summary
# ----------------------------------------------

print("===== F1 Analytics Environment Configuration =====")
print(f"BASE_PATH:        {BASE_PATH}")
print(f"LANDING_PATH:     {LANDING_PATH}")
print(f"BRONZE_PATH:      {BRONZE_PATH}")
print(f"SILVER_PATH:      {SILVER_PATH}")
print(f"GOLD_PATH:        {GOLD_PATH}")
print(f"INCREMENTAL_PATH: {INCREMENTAL_PATH}")
print(f"CONTROL_PATH:     {CONTROL_PATH}")
print("===================================================")
```
