```python
from pyspark.sql import functions as F

def add_ingestion_metadata(df, source_path: str):
    """
    Adds ingestion timestamp and source file path.
    Local version (no _metadata.file_path).
    """
    return (
        df.withColumn("ingest_timestamp", F.current_timestamp())
          .withColumn("source_file", F.lit(source_path))
    )


def write_to_bronze(df, output_path: str, batch_id: str):
    """
    Writes Bronze data as Parquet, partitioned by batch_id.
    Equivalent to Databricks Delta overwrite + replaceWhere.
    """
    (
        df.withColumn("batch_id", F.lit(batch_id))
          .write
          .mode("overwrite")
          .partitionBy("batch_id")
          .parquet(output_path)
    )
```
